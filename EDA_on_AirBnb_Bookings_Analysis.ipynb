**Lets remove the outliers first:**

````python
# Function to remove outliers from 'price' variable using IQR method
def remove_outliers(data, column):
    q1 = df['price'].quantile(0.25)
    q3 = df['price'].quantile(0.75)
    IQR = q3-q1
    lower_bound = q1 - 1.5*IQR
    upper_bound = q3 + 1.5*IQR
    return data[(data[column]>= lower_bound ) & (data[column] <= upper_bound)]

# Getting a new dataset with outliers removed
filtered_data = remove_outliers(df, 'price')

#printing the new dataset
filtered_data
````

````python
# Box plot after the removal of outliers in price.

sns.boxplot(x='room_type', y ='price', data=filtered_data)
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
````

**We have removed the outliers, now the box plot is reasonably more informativel**
